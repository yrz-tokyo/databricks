# ノートブックの作成

**１、ホームフォルターの紹介：**各ユーザーごとのホームフォルダ。自分のアセットはデフォルトでホームフォルダ配下に作成されます。アクセス権を付与しない限り、他のユーザーがあなたのアセットにアクセスすることはできません。しかし、管理ユーザーはこの限りではありません。<br>
**２、Sharedの紹介：**すべてのユーザーが読み書きできるフォルダです。ノートブックの共有などの目的で使用します。<br>
**３、Workspacesの紹介：** ワークスペースに登録されたユーザーそれぞれのホームフォルダです、アクセス権限ある場合アクセス可能。<br>
**４、お気に入り：**ノートブックやフォルダをお気に入りに登録するとこちらに表示されます。ショートカットできます。<br>
**５、ゴミ箱：** 削除されたノートブックなどは一定期間こちらに格納されます。元に戻すこともできます。<br>
**６、ノートブックを作成してみる：**{date}_the_first_notebook
